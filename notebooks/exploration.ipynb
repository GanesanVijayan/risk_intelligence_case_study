````python

# try huggingface api provider for model inferencing

from asyncio.windows_events import NULL
import json
import os
from typing import Dict
from huggingface_hub import InferenceClient
from openai import OpenAI


HF_API_KEY = "hf_XXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXX"
MODEL_NAME = "Qwen/Qwen3.5-4B:featherless-ai"

client = OpenAI(base_url="https://router.huggingface.co/v1", api_key=HF_API_KEY,)

# # Initialize Hugging Face Inference Client
# client = InferenceClient(api_key=HF_API_KEY)

def summarize_and_classify(candidate: str, page_ref: str) -> Dict[str, str]:
    """
    Summarize and classify candidate risk text using Hugging Face Inference API.
    Handles both single-risk and multi-risk cases.
    """

    prompt = f"""
    You are an assistant that extracts structured risk information.

    Candidate risk text (from {page_ref}):
    "{candidate}"

    Instructions:
    - If the text contains ONE clear risk, return a single JSON object with:
        risk_title, description, category, page_reference, mitigation.
    - If the text contains MULTIPLE distinct risks, return a JSON array of up to FIVE objects,
      each with the same keys (risk_title, description, category, page_reference, mitigation).
    - risk_title: short title (max 6 words).
    - description: 2–3 sentence summary.
    - category: one of [financial, operational, regulatory, market, climate, cyber, supply chain].
    - page_reference: {page_ref}.
    - mitigation: mitigation actions if mentioned, else null.
    """
    structured = None
    try:
        completion = client.chat.completions.create(
        model=MODEL_NAME,
        messages=[{"role": "user", "content": prompt}],)

        response = completion.choices[0].message.content
        structured = json.loads(response)
    except Exception as e:
        if "Client error '402 Payment Required'" in str(e):
            print("API quota exceeded. Please check your Hugging Face account and API usage.")
        else:
            print(f"Error during model inference: {e}")
            # fallback if model output is not valid JSON
            structured = [{
                "risk_title": None,
                "description": None,
                "category": None,
                "page_reference": None,
                "mitigation": None,
            }]

    return structured

````

````python
import json

def filter_risks_by_page(data, page_number: str):
    """
    Filter risks from the consolidated report by page number.
    
    Args:
        data (list): Parsed JSON list of reports.
        page_number (str): Page reference to filter, e.g. "p.118".
    
    Returns:
        list: Risks matching the given page number.
    """
    filtered_risks = []
    for report in data:
        for risk in report.get("risks", []):
            if risk.get("page_reference") == page_number:
                filtered_risks.append(risk)
    return {"risks": filtered_risks}


# Example usage
if __name__ == "__main__":
    sample_json = """
    [
       {
          "company_name": "Vestas",
          "year": "2025",
          "file_name": "VestasAnnualReport2025.pdf",
          "risks": [
             {
                "risk_title": "Cyber Attacks on Vestas' Digital Infrastructure",
                "description": "Potential cyber incidents disrupting power production, outages, and financial losses due to politically/motivated actors targeting wind turbines and grid control systems.",
                "category": "cyber",
                "page_reference": "p.118",
                "mitigation": ["Strategic cyber risk management"]
             },
             {
                "risk_title": "Geopolitical Cyber Threats Affecting Vestas' Customers",
                "description": "Increased risk of cyber incidents due to geopolitical instability, targeting customers/producers via Vestas’ exposure as a potential attack vector.",
                "category": "cyber",
                "page_reference": "p.118",
                "mitigation": ["Sharing expertise in international organizations"]
             }
          ]
       },
       {
          "company_name": "Vestas",
          "year": "2025",
          "file_name": "VestasAnnualReport2025.pdf",
          "risks": [
             {
                "risk_title": "Operational Risks",
                "description": "Exposure to diverse operational challenges due to global business complexity.",
                "category": "operational",
                "page_reference": "p.50",
                "mitigation": []
             }
          ]
       }
    ]
    """


    data = json.loads(sample_json)
    result = filter_risks_by_page(data, "p.118")
    for risk in result["risks"]:
      unique_risk_categories1 = {risk["category"]}
    print(unique_risk_categories1)

    result = filter_risks_by_page(data, "p.118")
    for risk in result["risks"]:
      unique_risk_categories2 = {risk["category"]}
    print(unique_risk_categories2)

    if any(unique_risk_categories1 & unique_risk_categories2):
        print("There are common risk categories between the two pages.")
    else:
        print("There are no common risk categories between the two pages.")

````

````python
import json

def compare_categories_by_page(golden_set, output_set, page_number: str):
    """
    Compare categories of risks for a given page_reference between golden and output sets.
    
    Args:
        golden_set (list): Parsed JSON list of golden reports.
        output_set (list): Parsed JSON list of output reports.
        page_number (str): Page reference to filter, e.g. "p.118".
    
    Returns:
        dict: {
            "matched": [...],
            "missing_in_output": [...],
            "extra_in_output": [...]
        }
    """
    # Extract categories from golden set
    golden_categories = set()
    for report in golden_set:
        for risk in report.get("risks", []):
            if risk.get("page_reference") == page_number:
                golden_categories.add(risk.get("category"))

    # Extract categories from output set
    output_categories = set()
    for report in output_set:
        for risk in report.get("risks", []):
            if risk.get("page_reference") == page_number:
                output_categories.add(risk.get("category"))

    # Compare sets
    matched = list(golden_categories & output_categories)
    missing_in_output = list(golden_categories - output_categories)
    extra_in_output = list(output_categories - golden_categories)

    return {
        "matched": matched,
        "missing_in_output": missing_in_output,
        "extra_in_output": extra_in_output
    }


# Example usage
if __name__ == "__main__":
    # Load your golden and output sets from JSON files or strings
    with open("data/golden_set.json", "r") as f:
        golden_set = json.load(f)
    with open("outputs/VestasAnnualReport2025_consolidated_risk_report_20260622_202610.json", "r") as f:
        output_set = json.load(f)

    result = compare_categories_by_page(golden_set, output_set, "p.118")
    print(json.dumps(result, indent=2))
````

````python
# user querying from final risk report

import json
import pandas as pd
import regex
from openai import OpenAI
from tabulate import tabulate

# Load API key
def load_api_key(config_path: str = "conf/config.yaml") -> str:
    with open(config_path, "r") as f:
        config = json.load(f)
    return config.get("SK_LM_API_KEY")

SK_LM_API_KEY = load_api_key()
# MODEL_NAME = "mistralai/ministral-3-3b"
MODEL_NAME = "qwen2.5-0.5b-instruct"

# Initialize LM Studio client
client = OpenAI(base_url="http://127.0.0.1:1234/v1", api_key=SK_LM_API_KEY)

def clean_and_parse_json(raw_text: str):
    """Remove markdown fences and parse JSON safely."""
    cleaned = regex.sub(r"```json|```", "", raw_text).strip()
    match = regex.search(r"\[.*\]|\{.*\}", cleaned, regex.DOTALL)
    if match:
        cleaned = match.group(0)
    try:
        return json.loads(cleaned)
    except json.JSONDecodeError as e:
        print("JSON parsing failed:", e)
        print("Cleaned text was:\n", cleaned)
        return None

def query_report(user_query: str, report_content: str):
    """
    Send user query + report content to LLM and return structured JSON.
    """
    prompt = f"""
    You are an assistant that applies filters and fetches results from a risk intelligence report.

    User query:
    "{user_query}"

    Report content:
    {report_content}

    Instructions:
    - Report content is in json format. Read it as json structured data
    - Identify the filter criteria from the user query (e.g., category, page_reference, risk_title).
    - Apply the filter strictly: return ONLY the risks that match the filter criteria.
    - Do not include risks that do not match the filter.
    - Output must be a JSON array of objects.
    - Each object must have: risk_title, description, category, page_reference, mitigation.
    - mitigation must be a list of strings (empty list if none).
    - Do not include explanations, commentary, or extra text outside the JSON.
    """

    completion = client.chat.completions.create(
        model=MODEL_NAME,
        messages=[{"role": "user", "content": prompt}],
        temperature=0.1,
        max_tokens=8192,
    )

    response = completion.choices[0].message.content.strip()
    structured = clean_and_parse_json(response)
    return structured

def to_pandas_table(structured_json):
    """
    Convert structured JSON results to pandas DataFrame for readability.
    """
    if structured_json:
        df = pd.DataFrame(structured_json)
        return df
    else:
        return pd.DataFrame()

# Example usage
if __name__ == "__main__":
    # Load consolidated risk report JSON
    with open("outputs/VestasAnnualReport2025_consolidated_risk_report_20260622_202610.json", "r") as f:
        report_content = f.read()

    # User query
    # user_queries = {"q1": "What are the top enterprise risks facing Vestas, and how frequently are they reviewed at Board/Executive level?",
    #                 "q2": "What emerging risks have been newly elevated?",
    #                 "q3": "Show me every renewable-energy company that lists cyber security as a principal risk"}

    user_queries = {"q1": "show only risk records with category as 'climate'"}

    from datetime import datetime
    ts = datetime.now().strftime("%Y%m%d_%H%M%S")
    for key, user_query in user_queries:
        # Query the model
        structured_results = query_report(user_query, report_content)
        print(structured_results)

        # Convert to pandas table
        df = to_pandas_table(structured_results)
        tabulate_df = tabulate(df, headers='keys', tablefmt='grid')
        op_file = f"outputs\{key}_response_{ts}.txt"
        print(op_file)
        with open(op_file, "w") as file:
            file.write(tabulate_df)
````
